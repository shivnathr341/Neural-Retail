```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel("online_retail_II.xlsx")
```

```python
df.isnull().sum()/len(df)*100
```

```python
df=df.dropna(subset=['Customer ID'])
```

```python
df['Description']=df['Description'].fillna('Unknown')
```

```python
df.isnull().sum()
df.shape
```

```python
(df['Quantity']<=0).sum()
```

```python
(df['Price']<=0).sum()
```

```python
df=df[df['Quantity']>0]
```

```python
df=df[df['Price']>0]
```

```python
df.shape
```

```python
df.duplicated().sum()
```

```python
df=df.drop_duplicates()
```

```python
df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'])
```

```python
df['Total Price']=df['Quantity']*df['Price'] #Add a column of total price
```

```python
df['Total Price'].describe()
```

```python
q1=df['Quantity'].quantile(0.25)
q3=df['Quantity'].quantile(0.75)  #remove outliers
iqr=q3-q1
```

```python
upper_limit=q3+3*iqr
```

```python
outlier=df[df['Quantity']>upper_limit]
```

```python
len(outlier)
```

```python
df=df[df['Quantity']<=upper_limit]
```

```python
df.shape
```

```python
print(f" Date range -- {df['InvoiceDate'].min()} to {df['InvoiceDate'].max()} ")
```

```python
print(f" Unique customer id :: {df['Customer ID'].nunique()}")
```

```python
print(f" Total Revenue : ${df['Total Price'].sum() }")
```
